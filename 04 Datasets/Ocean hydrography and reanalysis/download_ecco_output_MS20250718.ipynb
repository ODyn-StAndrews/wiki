ECCOv4 release 4 output is available from the Physical Oceanography Distributed Active Archive Center or [PO.DAAC](https://podaac.jpl.nasa.gov/).  

The <strong style="color:blue;">ecco_access</strong> package has been written specifically to simplify data access and includes useful functions like:  

**ecco_podaac_to_xrdataset():** takes as input a text query or ECCO dataset identifier, plus start and end dates, and returns an xarray Dataset. Used in this notebook.  
**ecco_podaac_access():** takes the same input, but returns the URLs/paths or local files where the data is located  
<br>


<strong style="color:blue;">ecco_access</strong> is not yet available on `conda` or `pip`, so <span style="color:maroon;">this notebook sets it up so that it can be imported and used like any other Python package. Here we will use it to download ECCO output to the group workspace (`/gws/nopw/j04/co2clim/datasets/ECCOv4r4/`).</span>

# SECT. 2: Downloading ECCO Output
<hr style="border: 2px solid black;">


### (1) Specify the path to the 'ECCO-v4-Python-Tutorial' directory and import 'ecco_access'

````python
# E.g.:

# Add path to directory containing 'ecco_access'
import sys
sys.path.append('/gws/nopw/j04/co2clim/USERNAME/ECCO-v4-Python-Tutorial')

# Add ecco_access "package", and one other needed for this notebook
import ecco_access as ea
from os.path import join
````

### (2) Set output path to /Group Workspace/datasets/ECCOv4r4/

````python
# Specify where on gws to download to. 
# The download commands below will download all relevant .nc files and place them
#   in a sub-directory (named ShortName, see below) within my_download_dir/ECCO_V4r4_PODAAC/.
my_download_dir = '/gws/nopw/j04/co2clim/datasets/ECCOv4r4/'
````

### (3) Download (specify: ShortName and Start/EndDate)

ECCO output is organized into dataset identifiers or codes called "ShortNames". Find the ShortName for your variable of interest [here](https://ecco-v4-python-tutorial.readthedocs.io/ECCO_access_intro.html#Dataset-ShortNames-and-variables-associated-with-them).  

All output (.nc files, one per month) will be downloaded to `/my_download_dir/ECCO_V4r4_PODAAC/`, within a subfolder named 'ShortName'.  


````python
# download data and open xarray dataset
ShortName = 'ECCO_L4_OCEAN_VEL_05DEG_MONTHLY_V4R4'
ds = ea.ecco_podaac_to_xrdataset(ShortName,\
    StartDate='1992-01',EndDate='2017-12',\
    mode='download',\
    download_root_dir=join(my_download_dir,'ECCO_V4r4_PODAAC'))
ds
````

**<span style="color:maroon;">Troubleshooting:</span>**  

If you get an error saying `OSError: [Errno -51] NetCDF: Unknown file format: xxx.nc` about one of the files, just run the cell again and it should work.

# SECT. 3: Accessing ECCO Output
<hr style="border: 2px solid black;">


Once downloaded, output can be loaded into any notebook with `xarray`'s open-multiple-files command. Just point to the subdirectory of interest, e.g.: 

````python
import xarray as xr

ds = xr.open_mfdataset('gws/nopw/j04/co2clim/datasets/ECCOv4r4/ECCO_V4r4_PODAAC/ECCO_L4_OCEAN_VEL_05DEG_MONTHLY_V4R4/*.nc')
````
